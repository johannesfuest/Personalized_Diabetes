```python
from matplotlib import pyplot as plt
import numpy as np
import plotly.express as px

import plotly.graph_objects as go
```

```python

y = np.linspace(0,400,500)
y_hat = np.linspace(0,400,500)

Y , Y_hat = np.meshgrid(y, y_hat)

```

```python
def mse(y, y_hat):
    return (y-y_hat)**2

def mae(y, y_hat):
    return abs(y-y_hat)

# Create the 3D surface plot
fig = go.Figure(data=go.Surface(x=Y, y=Y_hat, z=mse(Y, Y_hat)))

# Set the plot layout
fig.update_layout(
    title='MSE',
    scene=dict(
        xaxis_title='x',
        yaxis_title='y',
        zaxis_title='f(x, y)'
    )
)

# Show the plot
fig.show()
```

```python
def light_on(y, y_hat):
    hyper =   np.where(((y>100) & (y_hat<100)), 1, 0)
    hypo =    np.where(((y<70) & (y_hat>70)), 1, 0)
    
    return (hyper+hypo )

# Create the 3D surface plot
fig = go.Figure(data=go.Surface(x=Y, y=Y_hat, z=light_on(Y, Y_hat)))

# Set the plot layout
fig.update_layout(
    title='MSE',
    scene=dict(
        xaxis_title='y',
        yaxis_title='y_hat',
        zaxis_title='f(y, h_hat)'
    )
)

# Show the plot
fig.show()
```

```python
def sigmoid(y, y_hat, upper=100, lower=70, softener=2):
    loss_upper = np.minimum(1/(1 + np.exp((y_hat-upper)/softener)), 1/(1 + np.exp((-y+upper)/softener)))
    loss_lower = np.minimum(1/(1 + np.exp((-y_hat+lower)/softener)), 1/(1 + np.exp((y-lower)/softener)))
    return 1*(loss_lower + loss_upper) * mse(y, y_hat) + 0.5*mse(y, y_hat)

# Create the 3D surface plot
fig = go.Figure(data=go.Surface(x=Y, y=Y_hat, z=sigmoid(Y, Y_hat)))

# Set the plot layout
fig.update_layout(
    title='MSE',
    scene=dict(
        xaxis_title='y',
        yaxis_title='y_hat',
        zaxis_title='f(y, h_hat)'
    )
)

# Show the plot
fig.show()
```

 let's reimplement their work as it's well-published and has been cited quite often already.


 Please note that this is the **second attempt**. The first time, I corrected the sigmoid function by moving the boundaries of the function elements but this lead to inconsistent results with the main part. 
 Now, the result is consistent with the numbers provided in III-new metric. 

```python
def xi(x, a, epsilon):
    return 2/epsilon * (x-a-2/epsilon)

def xi_corrected(x, a, epsilon):
    return 2/epsilon * (x-a-epsilon/2)

def sigmoid(x, a, epsilon):
    XI = xi(x, a, epsilon)
    #print(XI)
    return np.where(
        x<=a,
        0,
        np.where(
            x <= a + epsilon/2,
            -1/2 *  (XI**4) - (XI**3) + XI + 1/2,
            np.where(
                x <= a + epsilon,
                1/2 * XI**4 - XI**3 + XI + 1/2,
                1
            )
        )
    )


# corrected
def sigmoid_corrected(x, a, epsilon):
    XI = xi_corrected(x, a, epsilon)
    #print(XI)
    return np.where(
        x<=a,
        0,
        np.where(
            x <= a + epsilon/2,
            -1/2 *  (XI**4) - (XI**3) + XI + 1/2,
            np.where(
                x <= a + epsilon,
                1/2 * XI**4 - XI**3 + XI + 1/2,
                1
            )
        )
    )

b = np.linspace(0,500,500)
fig = px.scatter(x=b, y= sigmoid(b, 155, 100), title='original')
fig.show()
fig = px.scatter(x=b, y=sigmoid_corrected(b, 155, 100), title='corrected')
fig.show()

```

```python
def xi_neg(x, a, epsilon):
    return -2/epsilon * (x-a+2/epsilon)

def xi_neg_corrected(x, a, epsilon):
    return -2/epsilon * (x-a+epsilon/2)

def sigmoid_neg(x, a, epsilon):
    XI = xi_neg(x, a, epsilon)
    #print(XI)
    return np.where(
        x<=a-epsilon,
        1,
        np.where(
            x <= a - epsilon/2,
            1/2 *  (XI**4) - (XI**3) + XI + 1/2,
            np.where(
                x <= a ,
                -1/2 * XI**4 - XI**3 + XI + 1/2,
                0
            )
        )
    )


# corrected
def sigmoid_neg_corrected(x, a, epsilon):
    XI = xi_neg_corrected(x, a, epsilon)
    #print(XI)
    return np.where(
        x<=a-epsilon,
        1,
        np.where(
            x <= a - epsilon/2,
            1/2 *  (XI**4) - (XI**3) + XI + 1/2,
            np.where(
                x <= a ,
                -1/2 * XI**4 - XI**3 + XI + 1/2,
                0
            )
        )
    )

b = np.linspace(0,500,1000)
fig = px.scatter(x=b, y= sigmoid_neg(b, 85, 30), title='original')
fig.show()
fig = px.scatter(x=b, y=sigmoid_neg_corrected(b, 85, 30), title='corrected')
fig.show()

```

## Putting the loss function together

```python
alpha_L, alpha_H, beta_L, beta_H, gamma_L, gamma_H, t_L, t_H = 1.5, 1, 30, 100, 10, 20, 85, 155

def Pen(g, g_hat, alpha_L = alpha_L, alpha_H = alpha_H, beta_L = beta_L, beta_H = beta_H, gamma_L = gamma_L, gamma_H = gamma_H, t_L = t_L, t_H = t_H):
    return 1 + alpha_L * sigmoid_neg_corrected(g, t_L, beta_L) * sigmoid_corrected(g_hat, g, gamma_L) + alpha_H * sigmoid_corrected(g, t_H, beta_H) * sigmoid_neg_corrected(g_hat, g, gamma_H)

def gSE(g, g_hat, alpha_L = alpha_L, alpha_H = alpha_H, beta_L = beta_L, beta_H = beta_H, gamma_L = gamma_L, gamma_H = gamma_H, t_L = t_L, t_H = t_H): 

    return np.square(g - g_hat) * Pen(g, g_hat, alpha_L, alpha_H, beta_L, beta_H, gamma_L, gamma_H, t_L, t_H)
    
def gMSE(g, g_hat, alpha_L = alpha_L, alpha_H = alpha_H, beta_L = beta_L, beta_H = beta_H, gamma_L = gamma_L, gamma_H = gamma_H, t_L = t_L, t_H = t_H):
    np.mean(gSE(g, g_hat, alpha_L, alpha_H, beta_L, beta_H, gamma_L, gamma_H, t_L, t_H))
```
